NOW THAT WE HAVE DONE EDA ON THE DATA SET LETS MOVE ON THE CREATING THE MODEL THAT PREDICTS IF YOU ARE LIKELY TO HAVE A HEART ATTCK OR NOT

```python
df.head()
```

```python
x = df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2 , random_state = 69 , shuffle = True)
```

```python
sc = StandardScaler()
```

```python
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
logr = LogisticRegression()
```

```python
logr.fit(xtrain,ytrain)
```

```python
yhat = logr.predict(xtest)
yhat
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(ytest,yhat)
cm
```
